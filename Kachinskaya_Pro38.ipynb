```python
import pandas as pd

events_df = pd.read_csv('data/7_4_Events.csv', sep=',')
purchase_df = pd.read_csv('data/purchase.csv', sep=',')

cond = (events_df.start_time>='2018-01-01') & (events_df.start_time<'2019-01-01') & (events_df.event_type=='registration')
registered = events_df[cond]['user_id'].to_list() # список пользователей, зарег. в 2018
events = events_df[events_df.user_id.isin(registered)]
events.start_time = pd.to_datetime(events.start_time, format='%Y-%m-%dT%H:%M:%S')

purchase=purchase_df[purchase_df.user_id.isin(registered)]


medium_users_count = events[events["selected_level"] == "medium"]["user_id"].nunique()
medium_users=events[events["selected_level"] == "medium"]["user_id"].to_list() #список пользователей с уровнем medium
print(medium_users_count)

hard_users_count = events[events["selected_level"] == "hard"]["user_id"].nunique()
hard_users=events[events["selected_level"] == "hard"]["user_id"].to_list() #список пользователей с уровнем hard
print(hard_users_count)

easy_users_count = events[events["selected_level"] == "easy"]["user_id"].nunique()
easy_users=events[events["selected_level"] == "easy"]["user_id"].to_list() #список пользователей с уровнем easy
print(easy_users_count)


level_choice_users_count = events[events["event_type"] == "level_choice"]["user_id"].nunique()
print(level_choice_users_count-medium_users_count-hard_users_count-easy_users_count) #убедились что выбраны все польз с уровнями

paying_users_count = purchase["user_id"].nunique()
print(paying_users_count)

registered_users_count = events[events["event_type"] == "registration"]["user_id"].nunique() #количество зарегистрированных 

purchase=purchase_df[purchase_df.user_id.isin(registered)]
purchase.event_datetime = pd.to_datetime(purchase.event_datetime, format='%Y-%m-%dT%H:%M:%S')

medium=purchase[purchase.user_id.isin(medium_users)]
paying_medium_user_count=medium['user_id'].nunique()
percent_medium_reg=paying_medium_user_count/registered_users_count
percent_medium=paying_medium_user_count/paying_users_count
print(
    "Процент пользователей, выбравших уровень medium (от общего числа зарегистрировавшихся): {:.2%}".format(
        percent_medium_reg
    ))

print(
    "Процент пользователей, выбравших уровень medium (от общего чмсла оплативших): {:.2%}".format(
        percent_medium
    )
)

hard=purchase[purchase.user_id.isin(hard_users)]
paying_hard_user_count=hard['user_id'].nunique()
percent_hard_reg=paying_hard_user_count/registered_users_count
percent_hard=paying_hard_user_count/paying_users_count
print(
    "Процент пользователей, выбравших уровень hard (от общего числа зарегистрировавшихся): {:.2%}".format(
        percent_hard_reg
    ))

print(
    "Процент пользователей, выбравших уровень hard (от общего чмсла оплативших): {:.2%}".format(
        percent_hard
    )
)


easy=purchase[purchase.user_id.isin(easy_users)]
paying_easy_user_count=easy['user_id'].nunique()
percent_easy_reg=paying_easy_user_count/registered_users_count
percent_easy=paying_easy_user_count/paying_users_count
print(
    "Процент пользователей, выбравших уровень easy (от общего числа зарегистрировавшихся): {:.2%}".format(
        percent_easy_reg
    ))

print(
    "Процент пользователей, выбравших уровень easy (от общего чмсла оплативших): {:.2%}".format(
        percent_easy
    )
)
level_df = events[events['event_type'] == 'level_choice']
level_medium = level_df[level_df['selected_level'] == 'medium']
level_hard = level_df[level_df['selected_level'] == 'hard']
level_easy = level_df[level_df['selected_level'] == 'easy']

purchase_medium = purchase[purchase.user_id.isin(medium_users)]
purchase_hard = purchase[purchase.user_id.isin(hard_users)]
purchase_easy = purchase[purchase.user_id.isin(easy_users)]

merged_medium = level_medium.merge(
    purchase_medium, on="user_id", how="inner"
)
merged_medium["timedelta"] = (
    merged_medium["event_datetime"] - merged_medium["start_time"]
)
print(merged_medium["timedelta"].mean())

merged_hard = level_hard.merge(
    purchase_hard, on="user_id", how="inner"
)
merged_hard["timedelta"] = (
    merged_hard["event_datetime"] - merged_hard["start_time"]
)
print(merged_hard["timedelta"].mean())

merged_easy = level_easy.merge(
    purchase_easy, on="user_id", how="inner"
)
merged_easy["timedelta"] = (
    merged_easy["event_datetime"] - merged_easy["start_time"]
)
print(merged_easy["timedelta"].mean())
```
